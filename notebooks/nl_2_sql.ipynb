###  Synthèse des données NBA

Ce projet utilise un fichier Excel composé de **trois feuilles complémentaires** permettant d’analyser les performances des joueurs NBA.

---

#### 1. `players` — Données principales des joueurs  
- **569 lignes × 45 colonnes**  
- **1 ligne = 1 joueur**  
- Contient les statistiques individuelles et avancées :  
  - Informations générales : `Player`, `Team`, `Age`  
  - Statistiques de base : `GP`, `W`, `L`, `Min`, `PTS`, `FGM`, `FGA`, `FG%`  
  - Statistiques avancées : `OREB%`, `DREB%`, `REB%`, `EFG%`, `TS%`, `USG%`, `PACE`, `PIE`, `POSS`  
- C’est la table principale utilisée pour l’EDA.

---

#### 2. `teams` — Référentiel des équipes  
- **30 lignes × 2 colonnes**  
- Colonnes :  
  - `Code` : code équipe (ex : ATL, BOS, LAL)  
  - `Nom complet de l'équipe` : nom officiel de la franchise  
- Permet d’enrichir `players` via une jointure sur le code équipe.

---

#### 3. `descriptions` — Dictionnaire de données  
- **45 lignes × 2 colonnes**  
- Colonnes :  
  - `Data` : nom de la variable  
  - `Descriptions` : définition de la variable  
- Correspond exactement aux 45 colonnes de `players`.  
- Sert de documentation pour comprendre chaque statistique.

---

#### Relations entre les tables  
- `players.Team` ↔ `teams.Code`  
  -> permet d’ajouter le nom complet des équipes  
- `players.columns` ↔ `descriptions.Data`  
  -> permet d’associer une définition à chaque variable

---

#### Structure de notre base de données
- `teams.Code` -> clé primaire
- `players.Team` -> clé étrangère
- `descriptions.Data` -> clé primaire

```python
# comme nous avons beaucoups de colonne dans players (nous allons essayé de gérer la creation de la base automatiquement)
from sqlalchemy import Column, Integer, Float, String

# Nous allons écrire un scripts qui lit notre DataFrame
# détecte le type de chaque colonne et génère automatiquement le code SQLAlchemy correspondant

def sqlalchemy_type(dtype):
    if "int" in str(dtype):
        return "Integer"
    if "float" in str(dtype):
        return "Float"
    return "String"

print("class Player(Base):")
print("    __tablename__ = 'players'")
print("    id = Column(Integer, primary_key=True, autoincrement=True)")

for col, dtype in players.dtypes.items():
    col_name = col.replace(" ", "_").replace("%", "pct").replace("+/-", "plus_minus").replace(":", "_")
    sql_type = sqlalchemy_type(dtype)
    print(f"    {col_name} = Column({sql_type})")
```

## Ingestion des Données dans la DB

```python
# j'ai crée la base de donnée avec le fichier create_db.players
# chargons nos différents données
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
```

```python
import os, sys
sys.path.append(os.path.abspath('..'))
from utils.create_db import Team, Description, Player, Base
```

```python
engine = create_engine("sqlite:///../data/nba.db")
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# Insertion teams
for _, row in teams.iterrows():
    session.add(Team(Code=row["Code"], TeamName=row["Nom complet de l'équipe"]))
```

```python
descriptions = descriptions.astype(str)
# Insertion descriptions
for _, row in descriptions.iterrows():
    session.add(Description(Data=row["Data"], Description=row["Descriptions"]))
```

```python
# pour le fichier principal convertissons les clés du dataframe
column_mapping = {
    "FG%": "FGpct",
    "3P%": "_3Ppct",
    "3PA": "_3PA",
    "15:00:00": "_15_00_00",
    "+/-": "plus_minus",
    "FT%": "FTpct",
    "AST%": "ASTpct",
    "OREB%": "OREBpct",
    "DREB%": "DREBpct",
    "REB%": "REBpct",
    "EFG%": "EFGpct",
    "TS%": "TSpct",
    "USG%": "USGpct",
    "AST/TO": "AST_TO",
    "AST RATIO": "AST_RATIO",
    "TO RATIO": "TO_RATIO",
}

def convert_keys(data):
    new_data = {}
    for k, v in data.items():
        if k in column_mapping:
            new_data[column_mapping[k]] = v
        else:
            new_data[k] = v
    return new_data
```

```python
# Inerstion et remplisable des colonnes manquantes
for _, row in players.iterrows(): 
    data = row.to_dict() 
    clean_data = convert_keys(data) 
    session.add(Player(**clean_data))
```

```python
session.commit() 
session.close()
```

### Amélioration de l'agent

```python
# Pour une question de sécurité nous allons ajouté
# Une validation de la requête SQL avant éxecution
def validate_sql(query:str):
    q = query.lower().strip()
    if not q.startswith('select'):
        raise ValueError("Seules les requêtes SELECT sont autorisées.")
    if any(word in q for word in ["drop", "delete", "update", "insert", "alter"]):
        raise ValueError("Requête SQL potentiellement dangereuse.")
    return query

# si aussi la requête SQL échoue
def safe_execute(query: str):
    try:
        return execute_query.run(query)
    except Exception as e:
        return f"ERREUR_SQL: {str(e)}"
```

```python
sql_only_prompt = PromptTemplate.from_template(
f"""
Tu es un expert SQL spécialisé en SQLite et en analyse de données NBA.

Ta mission :
- Lire la question utilisateur.
- Générer une requête SQL SQLite valide.
- Utiliser uniquement les tables et colonnes présentes dans le schéma.
- Ne renvoyer QUE la requête SQL, sans texte autour.

Voici les tables disponibles et leurs schémas :
{{table_info}}

Nombre maximum de lignes à renvoyer : {{top_k}}

Voici quelques exemples (few-shots) pour t’aider à comprendre le style attendu :

{few_shoots}

Règles importantes :
- Utilise uniquement les colonnes présentes dans le schéma.
- Pas de texte hors SQL.
- Pas de commentaires.
- Pas de guillemets inutiles.
- Pas de LIMIT 1 inutile.
- Si la question demande un top N, utilise ORDER BY + LIMIT.
- Si la question mentionne une équipe, utilise la colonne Team (code à 3 lettres).
- Si la question mentionne un joueur, utilise la colonne Player (nom complet exact).

Question utilisateur :
{{input}}

SQLQuery:
"""
)
```

```python
generate_sql = create_sql_query_chain(
    llm=llm,
    db=db,
    prompt=sql_only_prompt
)
sql = generate_sql.invoke({"question": "Donne-moi les 5 meilleurs rebondeurs"})
print(sql)
```

```python
# le prompt de sortie final
answer_prompt = PromptTemplate.from_template(
"""
Tu es un assistant expert en analyse de données NBA.

Ton rôle est d’interpréter des résultats SQL et de produire une réponse claire,
naturelle et utile pour l’utilisateur.

### Règles obligatoires :
1. Lis attentivement la question de l’utilisateur.
2. Lis la requête SQL générée (ne jamais la montrer dans la réponse).
3. Lis le résultat SQL exécuté.
4. Réponds uniquement à partir du résultat fourni.
5. Si le résultat est vide, dis-le simplement et propose une reformulation possible.
6. Si le résultat contient une erreur (ex : commence par "ERREUR_SQL"), explique l’erreur
   de manière simple et propose une nouvelle question correcte.
7. La réponse doit être en français, concise, et orientée “insight NBA”.
8. Ne jamais inventer de données. Ne jamais extrapoler au-delà du résultat.

### Informations fournies :

Question utilisateur :
{question}

Requête SQL générée (ne pas afficher dans la réponse) :
{query}

Résultat SQL :
{result}

### Réponse finale (en français, claire, concise, sans SQL) :
"""
)
rephrase_answer = answer_prompt | llm | StrOutputParser()
```

```python
nl_2_sql = (
    RunnablePassthrough.assign(query=generate_sql) # on genère la requête SQL
    .assign(query=lambda x: validate_sql(x["query"])) # on vérifie la requête SQL
    .assign(result=lambda x: safe_execute(x["query"])) # on exécute la requête
    | rephrase_answer  # puis on reformule la réponse
)
```

```python
print(nl_2_sql.invoke({"question": "quel est le meilleur joueur et il joue dans quelle équipe"}))
```
